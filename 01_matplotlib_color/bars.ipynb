```python
import matplotlib.pyplot as plt
import numpy as np
```

# Vertical bar charts

**Purpose**: Comparing categories **OR** showing temporal trends in data with few (< 4) time points.

**matplotlib function**: [`bar(left, height)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar)

* `left`: The x coordinate(s) of the left sides of the bars.

* `height`: The height(s) of the bars.

```python
years = np.arange(2012, 2015)
values = [2, 5, 9]

plt.figure()

plt.bar(years, values)
;
```

**Commonly used parameters**:

* `color`: Set the color of the bars.

* `edgecolor`: Set the color of the lines on the edges of the bars.

* `width`: Set the width of the bars.

* `align`: Set the alignment of the bars, e.g., center them on the x coordinate(s).

* `label`: Set the label for the bar that will show up in the legend.

```python
#we can play with bars position (and use multiple bars)
years = np.arange(2012, 2015)
category1_values = [2, 5, 9]
category2_values = [7, 6, 3]

plt.figure()

plt.bar(years - 0.2,
        category1_values,
        color='blue',
        edgecolor='none',
        width=0.4,
        align='center',
        label='y1')

plt.bar(years + 0.2,
        category2_values,
        color='orange',
        edgecolor='none',
        width=0.4,
        align='center',
        label='y2')

plt.xticks(years, [str(year) for year in years])

plt.legend()
;
```

Use of colormap: categorical. Problems? Try to change!

```python
categories = ['A', 'B', 'C', 'D', 'E','F','G','H','I','L','M']
values = [7, 12, 4, 2, 9,4,5,3,9,11,5]
my_cmap = plt.get_cmap("tab10")
plt.figure()

plt.bar(np.arange(len(categories)), values, color=my_cmap.colors)

plt.xticks(np.arange(len(categories)),
           ['Category {}'.format(x) for x in categories], rotation='vertical')
;
```

```python
#color by value
my_cmap = plt.get_cmap("viridis")
rescale = lambda y: (y - np.min(y)) / (np.max(y) - np.min(y))
plt.bar(np.arange(len(categories)), values, color=my_cmap(rescale(values)))

plt.xticks(np.arange(len(categories)),
           ['Category {}'.format(x) for x in categories], rotation='vertical');

plt.savefig("temp")
```

# Horizontal bar charts

**Purpose**: Comparing categories.

**matplotlib function**: [`barh(bottom, width)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh)

* `bottom`: The y coordinate(s) of the bars.

* `width`: The width(s) of the bars.

```python
categories = ['A', 'B', 'C', 'D', 'E','F','G','H','I','L','M']
values = [7, 12, 4, 2, 9,4,5,3,9,11,5]
my_cmap = plt.get_cmap("tab20")
plt.figure()

plt.barh(np.arange(len(categories)), values, color=my_cmap.colors)
plt.yticks(np.arange(len(categories)),
           ['Category {}'.format(x) for x in categories])
;
```

# stacked bar charts

**Purpose**: Comparing values with two categories


```python
# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```
