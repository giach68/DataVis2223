#### Importing Libraries and Data

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

t_mov= pd.read_csv('data/types_movies.csv')
```

# Pie charts

**Purpose**: Displaying a simple proportion.

**matplotlib function**: [`pie(sizes)`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie)

* `sizes`: The size of the wedges as either a fraction or number.

```python
plt.rcdefaults()

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'])

plt.axis('equal')

plt.show()
```

### Pie chart components
* <b>wedges:</b> A list of Patch objects representing each wedge
* <b>texts:</b> List of Text objects representing the labels
* <b>autotexts:</b> List of Text objects for the numeric values - this is only availabe if the autopct value for the pie chart is not None

```python
wedges, texts, autotexts = plt.pie(t_mov['Percentage'], 
                                    labels=t_mov['Sector'],
                                    autopct='%.2f')
plt.axis('equal')

print('Wedges: ', wedges)
print('Texts: ', texts)
print('Autotexts: ', autotexts)
```

**Commonly used parameters**:

* `colors`: Set the colors of the wedges.

* `labels`: Set the labels of the wedges.

* `startangle`: Set the angle that the wedges start at.

* `autopct`: Set the percentage display format of the wedges.
    

#### Pie chart customizations
 We set the values for the colors and autopct properties. The latter sets the format for the values to be displayed. 

```python
colors = ['darkorange', 'sandybrown', 'darksalmon', 'orangered','chocolate']

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f')

plt.axis('equal')

plt.show()
```

```python
#changing color with cycler
from cycler import cycler
plt.rcParams['axes.prop_cycle'] = cycler(color=['r', 'g', 'b', 'y','c'])
wedges, texts, autotexts = plt.pie(t_mov['Percentage'], 
                                    labels=t_mov['Sector'],
                                    autopct='%.2f')
plt.axis('equal')

plt.show()
plt.rcdefaults()
```

#### The explode property
To highlight a particular wedge of the pie chart, we use explode to separate it from the rest of the chart. 

The value for "explode" represents the fraction of the radius with which to offset each wedge.

```python
explode = (0, 0.1, 0, 0, 0.3)

plt.pie(t_mov['Percentage'], 
        labels=t_mov['Sector'], 
        colors=colors, 
        autopct='%.2f',
        explode=explode)

plt.axis('equal')

plt.show()
```
